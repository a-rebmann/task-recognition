# Evaluation results

```python
import warnings
warnings.filterwarnings('ignore')
res_file = "../data/output/results.csv"

import pandas as pd
import matplotlib.pyplot as plt
```

### Overall results

```python
df = pd.read_csv(res_file, sep=";")
df
```

```python
df_ours = df[df["config"].str.contains("warm_up_events=250")]
df_ours = df_ours.groupby("log").first().reset_index()
print(df_ours[["log", "Seg. true", "Seg. found", "ED", "rand_mic", "rand_mac", "jaccard_mic", "jaccard_mac"]].to_latex(index_names=False, index=False, float_format="%.2f"))
```

### Label-assignment results

```python
import pandas as pd
res_file = "../data/output/results.csv"
df = pd.read_csv(res_file, sep=";")
df_ours = df[(df["config"].str.contains("warm_up_events=offline")) & df["with_labels"]]
df_ours = df_ours.groupby("log").first().reset_index()
print(df_ours[["log", "label_prec", "label_rec", "label_f1", "label_support"]].to_latex(index_names=False, index=False, float_format="%.2f"))
```

### Object-identification results

```python
import pandas as pd
res_file = "../data/output/results.csv"
df = pd.read_csv(res_file, sep=";")
df_ours = df[df["config"].str.contains("warm_up_events=250")]
df_ours["count"] = df_ours["obj_tp"] + df_ours["obj_fn"]
df_ours = df_ours.groupby("log").first().reset_index()
print(df_ours[["log", "obj_prec", "obj_rec", "obj_f1"]].to_latex(index_names=False, index=False, float_format="%.2f"))
```
